```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_csv(file_name):
    df = pd.read_csv(f"Artista_data/{file_name}.csv", index_col=0)
    return df
```

```python
taylor_swift_df = read_csv('taylor_swift_spotify')
ed_sheeran_df = read_csv('ed_sheeran_spotify')
beatles_df = read_csv('beatles_spotify')
elton_john_df = read_csv('elton_john_spotify')
metallica_df = read_csv('metallica_spotify')
rolling_stones_df = read_csv('rolling_stones_spotify')
tame_impala_df = read_csv('tame_impala_spotify')
```

```python
def plot_liveness_over_carreer(df, artist):
    df['release_year'] = df['release_date'].apply(lambda x: int(x[:4]))
    grouped_df = df[['liveness', 'release_year']].groupby('release_year').mean()
    plt.fill_between(grouped_df.index, grouped_df['liveness'].rolling(2).mean())
    plt.yticks(np.arange(0,1.1,0.2))
    plt.show()
    grouped_df.rename(columns={'liveness': f"{artist}_liveness"}, inplace=True)
    return grouped_df
```

```python
ts_df = plot_liveness_over_carreer(taylor_swift_df, "taylor_swift")
```

```python
es_df = plot_liveness_over_carreer(ed_sheeran_df, "ed_sheeran")
```

```python
bt_df = plot_liveness_over_carreer(beatles_df, "beatles")
```

```python
ej_df = plot_liveness_over_carreer(elton_john_df, "elton_john")
```

```python
mt_df = plot_liveness_over_carreer(metallica_df, "metallica")
```

```python
rs_df = plot_liveness_over_carreer(rolling_stones_df, "rolling_stones")
```

```python
ti_df = plot_liveness_over_carreer(tame_impala_df, "tame_impala")
```

```python
combined_df = pd.concat([ts_df, es_df, bt_df, ej_df, mt_df, rs_df, ti_df], axis=1)
combined_df = combined_df.sort_values('release_year')
combined_df = combined_df.fillna(0.0)
for col in combined_df.columns:
    combined_df[col] = combined_df[col].rolling(2).mean()
combined_df
```

```python
plt.figure(figsize=(15,10))
plt.stackplot(combined_df.index, combined_df['taylor_swift_liveness'], combined_df['ed_sheeran_liveness'], combined_df['beatles_liveness'], 
              combined_df['elton_john_liveness'], combined_df['metallica_liveness'], combined_df['rolling_stones_liveness'], 
              combined_df['tame_impala_liveness'], labels=["Taylor Swift", "Ed Sheeran", "Beatles", "Eltonn John", "Metallica", "Rolling Stones", "Tame Impala"], baseline='sym')
plt.legend(loc='upper left')
```
