```python
# Imports

import pandas as pd
from plotnine import *
```

```python
# Datasets

beatles_df = pd.DataFrame(pd.read_csv("./Artista_data/beatles_spotify.csv", sep=','))
beatles_df["artist"] = "The Beatles"

ed_sheeran_df = pd.DataFrame(pd.read_csv("./Artista_data/ed_sheeran_spotify.csv", sep=','))
ed_sheeran_df["artist"] = "Ed Sheeran"

elton_john_df = pd.DataFrame(pd.read_csv("./Artista_data/elton_john_spotify.csv", sep=','))
elton_john_df["artist"] = "Elton John"

metallica_df = pd.DataFrame(pd.read_csv("./Artista_data/metallica_spotify.csv", sep=','))
metallica_df["artist"] = "Metallica"

rolling_stones_df = pd.DataFrame(pd.read_csv("./Artista_data/rolling_stones_spotify.csv", sep=','))
rolling_stones_df["artist"] = "Rolling Stones"

tame_impala_df = pd.DataFrame(pd.read_csv("./Artista_data/tame_impala_spotify.csv", sep=','))
tame_impala_df["artist"] = "Tame Impala"

taylor_swift_df = pd.DataFrame(pd.read_csv("./Artista_data/taylor_swift_spotify.csv", sep=','))
taylor_swift_df["artist"] = "Taylor Swift"

artists_df = pd.concat([beatles_df, ed_sheeran_df, elton_john_df, metallica_df, rolling_stones_df, tame_impala_df, taylor_swift_df])
artists_df["release_date"] = pd.to_datetime(artists_df["release_date"], format='mixed')
artists_df['year'] = artists_df['release_date'].dt.year

artists_df
```

# Q8: Intervalo entre lançamento de álbuns faz alguma diferença?

```python
#Intervalo entre lançamento de álbuns faz alguma diferença? considerando a base utilizada

import pandas as pd
# Calculate the time difference between each album's release date
artists_df['album_release_gap'] = artists_df.groupby('artist')['release_date'].diff()

# Group the data by artist and calculate the average album release gap
grouped_df = artists_df.groupby('artist')['album_release_gap'].mean().reset_index()

# Convert the time difference to days
grouped_df['album_release_gap'] = grouped_df['album_release_gap'] / pd.Timedelta('1 day')

# Plot the average album release gap for each artist
ggplot(grouped_df, aes(x='artist', y='album_release_gap')) + \
  geom_bar(stat='identity') + \
  labs(title='Average Album Release Gap by Artist', x='Artist', y='Average Gap (days)') + \
  theme(axis_text_x=element_text(rotation=45, hjust=1),
        figure_size=(9, 6))
```

```python
import pandas as pd
import plotnine as p9

# Calculate the time difference between each album's release date
artists_df['album_release_gap'] = artists_df.groupby('artist')['release_date'].diff()

# Group the data by artist and calculate the average album release gap
grouped_df = artists_df.groupby('artist')['album_release_gap'].mean().reset_index()

# Convert the time difference to days
grouped_df['album_release_gap'] = grouped_df['album_release_gap'] / pd.Timedelta('1 day')

# Create a scatter plot of popularity vs. average album release gap
p9.ggplot(grouped_df, p9.aes(x='album_release_gap', y='popularity')) + \
  p9.geom_point() + \
  p9.labs(title='Popularity vs. Average Album Release Gap', x='Average Gap (days)', y='Popularity') + \
  p9.theme(axis_text_x=p9.element_text(rotation=45, hjust=1),
            figure_size=(9, 6))

# Create a boxplot of popularity by artist
p9.ggplot(artists_df, p9.aes(x='artist', y='popularity')) + \
  p9.geom_boxplot() + \
  p9.labs(title='Popularity by Artist', x='Artist', y='Popularity') + \
  p9.theme(axis_text_x=p9.element_text(rotation=45, hjust=1),
            figure_size=(9, 6))

# Create a line chart of popularity over time for each artist
p9.ggplot(artists_df, p9.aes(x='release_date', y='popularity', color='artist')) + \
  p9.geom_line() + \
  p9.labs(title='Popularity Over Time by Artist', x='Release Date', y='Popularity') + \
  p9.theme(axis_text_x=p9.element_text(rotation=45, hjust=1),
            figure_size=(9, 6))
```
